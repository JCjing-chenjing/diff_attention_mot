```python
import pandas as pd
import numpy as np 

def get_map(sheet_data):
    sheet_map = {}
    for i in range(len(sheet_data)):
        value = sheet_data.iloc[i, :]
        value =  value.values
        if value[1] is np.nan:
            k = value[0] 
        else:
            k = value[0] + "_" + value[1]
        sheet_map[k] = value[2]
    return sheet_map

excel_path = r"C:\Users\Administrator\Desktop\check.xlsx"
sheet1_data = pd.read_excel(excel_path,sheet_name="Sheet1")
sheet1_map = get_map(sheet1_data)

sheet2_data = pd.read_excel(excel_path,sheet_name="Sheet2")
sheet2_map = get_map(sheet2_data)

for k in sheet1_map:
    if k not in sheet2_map:
        print(k, "不在评估字段中")
    if sheet1_map[k] != sheet2_map[k] and ((not sheet1_map[k]  is np.nan) or  (not sheet2_map[k] is np.nan)):
        print(k, sheet1_map[k], "==>", sheet2_map[k])
```

```python
import math
def position(max_len, d_model):
    pe = torch.zeros(max_len, d_model)
    position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
    div_term = torch.exp(torch.arange(
        0, d_model, 2).float() * (-math.log(10000.0) / d_model))
    pe[:, 0::2] = torch.sin(position * div_term)
    pe[:, 1::2] = torch.cos(position * div_term)
    pe = pe.unsqueeze(0).transpose(0, 1)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def getPositionEncoding(seq_len, d, n=10000):
    P = np.zeros((seq_len, d))
    for k in range(seq_len):
        for i in np.arange(int(d/2)):
            denominator = np.power(n, 2*i/d)
            P[k, 2*i] = np.sin(k/denominator)
            P[k, 2*i+1] = np.cos(k/denominator)
    return P

P = getPositionEncoding(seq_len=6, d=4, n=100)
print(P)
```
